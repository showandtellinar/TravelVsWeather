```python
def haversine(lon1, lat1, lon2, lat2):
    """
    Calculate the great circle distance between two points 
    on the earth (specified in decimal degrees)
    """
    # convert decimal degrees to radians 
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])

    # haversine formula 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 

    # 6367 km is the radius of the Earth
    km = 6367 * c
    return km * 1000
```

```python
df['time'] = df.index
df['delta_seconds'] = (df['time']-df['time'].shift()).astype('timedelta64[s]').fillna(0)
df.drop("time", axis=1, inplace=True)
df.head()
```

```python
df['prevLong'] = (df['long'].shift())
df['prevLat'] = (df['lat'].shift())
df.head()
```

```python
df = df.dropna()
df['distance_meters'] = df.apply(lambda r: haversine(r.long, r.lat, r.prevLong, r.prevLat), 1)
df.head()
```

```python
miles_per_meter = 0.000621371
seconds_per_hour = 60*60
df['mph'] = df.apply(lambda r: r['distance_meters'] * miles_per_meter / r['delta_seconds'] * seconds_per_hour, 1)
df.head()
```

```python
#Only keep rows with mph less than or equal to 100(some are crazy high from bad data)
df = df[df.mph <= 100]
```

```python
#Only keep rows with detla_seconds less than 600(10 mintes) and greater than 10 seconds
df = df[(df['delta_seconds'] <= 600)]
df = df[df['delta_seconds'] > 10]
print len(df.index)
```

```python
Activity = namedtuple("Activity", ['StartTime', 'EndTime', 'AvgSpeed'])

def getSpeedPeriods(speedData, minMinutesBetween, speedRange):
    results = []
    activity_data = speedData[(speedData.mph > speedRange.l) & (speedData.mph < speedRange.h)]
    period_start = None
    last_period = None
    period_count = 0
    for i, row in activity_data.iterrows():
        mph = row['mph']
        if period_start is None:
            period_start = i
            last_period = i
            period_count += 1
            continue
        if i - last_period <= timedelta(minutes=minMinutesBetween):
            period_count += 1
            last_period = i
        else:
            if period_count >= 2:
                results.append(Activity(period_start, last_period, activity_data[period_start:last_period]['mph'].mean()))
            period_start = i
            last_period = i
            period_count = 1
    return results
```
